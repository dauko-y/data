```python
import pandas as pd
import bs4 as bs
from bs4 import BeautifulSoup
import requests
import pandas as pd
import json
import bs4 as bs
import urllib.request
import seaborn as sns
import matplotlib as mpl
```

```python
source = urllib.request.urlopen('https://en.wikipedia.org/wiki/List_of_presidents_of_the_United_States').read()
soup = bs.BeautifulSoup(source,'lxml')
```

```python
table = soup.find_all('table')
df = pd.read_html(str(table))[0]
```

```python
display(df)
```

```python
null_count = df.isna().sum()
print(null_count)
```

```python
df_copy = df.drop(columns=["Portrait", "Party[b][15]"])
```

```python
df_copy.head()
```

```python
df_copy[['FirstName', 'Surname']] = df_copy['Name(Birth–Death)'].str.split(' ', n=1, expand=True)
df_copy[['Surname', 'Birth-Death']] = df_copy['Surname'].str.split('(', n=1, expand=True)
df_copy[['Birth', 'Death']] = df_copy['Birth-Death'].str.split('–', n=1, expand=True)
df_copy[['Death', 'XXX']] = df_copy['Death'].str.split(')', n=1, expand=True)
```

```python
df_copy.head()
```

```python
df_copy2 = df_copy.drop(columns=["Name(Birth–Death)", "Birth-Death", "XXX"])
```

```python
df_copy2.head()
```

```python
df_copy2 = df_copy2.reindex(columns=['No.[a]', 'FirstName', 'Surname', 'Birth', 'Death', 'Term[14]', 'Party[b][15].1', 'Election', 'Vice President[16]'])
```

```python
df_copy2.head()
```

```python
df_copy2.rename(columns = {'Party[b][15].1': 'Parthy'})
df_copy2.rename(columns = {'No.[a]': 'PresidentNumber'})
```
